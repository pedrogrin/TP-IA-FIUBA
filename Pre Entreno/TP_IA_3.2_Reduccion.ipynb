Como nuestros datos no son lineales (lo podemos ver en la matriz de correlacion) usaremos UMAP

```python
import umap.umap_ as umap
```

```python
reducer = umap.UMAP(n_components = 3)
```

```python
reducer.fit(X_train)
```

```python
result_train = pd.DataFrame(reducer.transform(X_train), columns=['UMAP%i' % i for i in range(3)], index=X_train.index)
```

```python
X_train = pd.concat([X_train, result_train], axis = 1)
```

```python
result_train.head()
```

```python
X_train.head()
```

```python
result_test = pd.DataFrame(reducer.transform(X_test), columns=['UMAP%i' % i for i in range(3)], index=X_test.index)
```

```python
X_test = pd.concat([X_test, result_test], axis = 1)
```

```python
stdScaler = StandardScaler()
features = ['UMAP0', 'UMAP1', 'UMAP2']

x_a_escalar_train = X_train.loc[:,features].values
x_a_escalar_train = stdScaler.fit_transform(x_a_escalar_train)
x_train_data_escalada = pd.DataFrame(data = x_a_escalar_train, columns = features)
X_train['UMAP0'] = x_train_data_escalada['UMAP0']
X_train['UMAP1'] = x_train_data_escalada['UMAP1']
X_train['UMAP2'] = x_train_data_escalada['UMAP2']

x_a_escalar_test = X_test.loc[:,features].values
x_a_escalar_test = stdScaler.transform(x_a_escalar_test)
x_test_data_escalada = pd.DataFrame(data = x_a_escalar_test, columns = features)
X_test['UMAP0'] = x_test_data_escalada['UMAP0']
X_test['UMAP1'] = x_test_data_escalada['UMAP1']
X_test['UMAP2'] = x_test_data_escalada['UMAP2']
```
