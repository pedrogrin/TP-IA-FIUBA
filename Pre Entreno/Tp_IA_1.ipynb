Vemos que estan muy desequilibradas las categorias, lo cual es malo

Separamos en dos df según la clase porque así al hacer el split nos aseguramos que al menos tengamos bien distribuidos los datos en cada set.

```python
X_clase_0 = df_clase_0.drop("HeartDisease", axis=1)
X_clase_1 = df_clase_1.drop("HeartDisease", axis=1)
y_clase_0 = df_clase_0["HeartDisease"]
y_clase_1 = df_clase_1["HeartDisease"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train_0, X_test_0, y_train_0, y_test_0 = train_test_split(X_clase_0, y_clase_0, test_size=0.3, random_state=10)
X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(X_clase_1, y_clase_1, test_size=0.3, random_state=10)
```

```python
X_train_inicial = pd.concat([X_train_0, X_train_1], ignore_index=True)
X_test_inicial = pd.concat([X_test_0, X_test_1], ignore_index=True)

y_train = pd.concat([y_train_0, y_train_1], ignore_index=True)
y_test = pd.concat([y_test_0, y_test_1], ignore_index=True)
```
